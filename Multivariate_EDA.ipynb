##Some Multivariate EDA: 

```python
plt.figure(figsize=(15,7))
sns.barplot(x='preferred_foot',y='value_eur',data=df,hue='body_type')
```

* In right foot prefrence category players the players which are having maximum value has body type of Neymar.
* In left foot prefrence category players the players which are having maximum value has body type of Messi.

```python
plt.figure(figsize=(15,7))
sns.barplot(x='preferred_foot',y='age',data=df,hue='body_type')
```

* In right foot prefrence category's players the players who are eldest in the category has body type of Akinfenwa and youngest has lean body type.
* In left foot prefrence category players the players which are eldest has body type of Messi and youngest has lean body type.

```python
plt.figure(figsize=(15,7))
sns.boxplot(x='work_rate',y='potential',data=df,hue='preferred_foot')
```

* If player's attact/defence work rate is  high/Medium then there is no difference in player's potential either they prefer right or left foot.
*  If player's attact/defence work rate is  Medium/Low, High/Low, Medium/Medium, High/High, Low/Medium then their potential is slightly higher in casse if they prefer left foot than right foot in game.
*  If player's attact/defence work rate is Medium/High and Low/High then their potential is slightly higher in case if they prefer right foot than left foot in game.


```python
plt.figure(figsize=(15,7))
sns.barplot(x='international_reputation',y='pace',data=df,hue='preferred_foot')
```

* As the International reputaion of players increasess tha left foot preference also increases.
